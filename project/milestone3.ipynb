```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
superstore_data = pd.read_csv("/Users/theranmeadows/Desktop/Bellevue University/dsc630predictiveAnalytics/project/superstore - data - superstore - superstore.csv")
superstore_data.head()
```

```python
superstore_grouped = superstore_data.groupby("Category")
category_sales = superstore_grouped['Profit'].sum()
```

```python
cat_sales_data = [['Furniture', 285204.72], ['Office Supplies', 518473.83], ['Technology', 663778.73]]
category_sales_df = pd.DataFrame(cat_sales_data, columns=['Category', 'Total Profit'])
```

```python
category_sales_df.plot(x = 'Category', y = 'Total Profit', kind = 'bar')
plt.title("Total Proft by Category")
plt.xlabel('Category')
plt.ylabel("Total Profit")
```
