### Write a summary of your data and identify at least two questions to explore visually with your data.

The Aircraft Wildlife Strikes 1990-2023 data set is a data set that contains over 30 years of airplanes strikes that involved wildlife. It was compiled by the Federal Aviation Administration and contains data such as Date, Airport, Damage Level, and even the species of wildlife that was hit.

Two questions I would like to answer about this data set is:
1) What was the most common species among the airplane strikes?
2) What year had the most strikes?

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the data
crash_data = pd.read_csv('/Users/theranmeadows/Desktop/dsc630predictiveAnalytics/week1/STRIKE_REPORTS.csv')
crash_data.columns
```

```python
crash_data.head()
```

### Create a histogram or bar graph from your data.

```python
count = crash_data['SPECIES'].value_counts().head(10)
count.head()
```

```python
count.plot(kind = 'bar')
plt.title('Species')
plt.xlabel('Species')
plt.ylabel('Number of Strikes')
plt.show()
```

### Create a boxplot from your data.

```python
plt.boxplot(crash_data['INCIDENT_YEAR'])
plt.title('Strikes by Year')
plt.show()
```

### Create a bivariate plot from your data.

```python
crash_data.plot(kind = 'scatter', x = 'INCIDENT_YEAR', y = 'NR_INJURIES')
```

### Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
plt.hist(crash_data['INCIDENT_YEAR'])
plt.title('Year')
plt.xlabel('Year')
plt.ylabel('Count of Strikes')
plt.show()
```
